```python
#import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Create Classifier
face_classifier=cv2.CascadeClassifier("I:/Haarcascades/haarcascade_frontalface_default.xml")
eye_classifier=cv2.CascadeClassifier("I:/Haarcascades/haarcascade_eye.xml")
```

```python
#Open and fix image
img=cv2.imread("I:/Images/eye_face.jpg")
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
#Detect Faces
faces=face_classifier.detectMultiScale(img,1.3,5)
```

```python
if faces is ():
    print("No Faces Found")
```

```python
def detect_face_eyes(fix_img):
    face_rects=face_classifier.detectMultiScale(fix_img)
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                      (x+w,y+h),
                      (255,0,0),
                      10)
        eyes_rects=eye_classifier.detectMultiScale(fix_img)
        for(ix,iy,iw,ih) in eyes_rects:
            cv2.rectangle(fix_img,
             (ix,iy),
            (ix+iw,iy+ih),
            (255,255,255),
            10)
    return fix_img
```

```python
result=detect_face_eyes(fix_img)
plt.imshow(result)
```
